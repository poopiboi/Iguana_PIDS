```python
import matplotlib.pyplot as plt

# Sample points
x = list(range(1, 11))
y_byhand = [3, 2, 2, 5, 1, 2, 1, 2, 2, 2]
y_model1 = [1, 3, 1, 2, 1, 2, 2, 1, 2, 2]
# Calculate absolute difference between the two sets of points
difference = [abs(byhand - model) for byhand, model in zip(y_byhand, y_model1)]

# Calculate sum of absolute difference
sum_difference = sum(difference)

# Plot difference
plt.plot(x, difference, "o", color = 'royalblue', label='Absolute Difference')  
plt.xlabel('Image_id')
plt.ylabel('Absolute Difference')
plt.title('Difference between By Hand and Model Labeling(25)')
#plt.grid(True)
plt.legend()

# Display sum of absolute difference
plt.text(3.7, max(difference) * -0.3, f'Sum of Absolute Difference: {sum_difference}', fontsize=10)

# Change y-axis tick labels to display only full numbers
plt.yticks(range(int(max(difference)) + 1))

# Display plot and sum of absolute difference underneath
plt.tight_layout()
plt.show()
```
